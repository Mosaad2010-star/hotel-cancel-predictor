# Dataset: **Hotel Bookings**

This dataset contains historical booking data from two types of hotels: **Resort Hotel** and **City Hotel**. It includes customer details, booking information, pricing, cancellation status, and reservation channels.

---

##  General Info:
- **Rows:** ~119,000 bookings  
- **Columns:** 32 features  
- **Target Variable:** `is_canceled` (1 = canceled, 0 = not canceled)

---

##  Feature Descriptions:

| Feature Name                  | Description |
|------------------------------|-------------|
| `hotel`                      | Type of hotel (Resort Hotel or City Hotel) |
| `is_canceled`                | Whether the booking was canceled |
| `lead_time`                  | Number of days between booking and check-in |
| `arrival_date_year`          | Arrival year |
| `arrival_date_month`         | Arrival month (string) |
| `arrival_date_week_number`   | Week number of the year |
| `arrival_date_day_of_month`  | Day of the month |
| `stays_in_weekend_nights`    | Nights stayed during weekends |
| `stays_in_week_nights`       | Nights stayed during weekdays |
| `adults`                     | Number of adults |
| `children`                   | Number of children |
| `babies`                     | Number of babies |
| `meal`                       | Type of meal plan |
| `country`                    | Guest country (ISO code) |
| `market_segment`             | Source of booking (online, group, etc.) |
| `distribution_channel`       | How the booking was distributed |
| `is_repeated_guest`          | Whether the guest is a returning customer |
| `previous_cancellations`     | Number of previous cancellations |
| `previous_bookings_not_canceled` | Number of successful past bookings |
| `reserved_room_type`         | Room type reserved |
| `assigned_room_type`         | Room type assigned |
| `booking_changes`            | Number of changes to the booking |
| `deposit_type`               | Deposit policy |
| `agent`                      | ID of the travel agent who made the booking |
| `company`                    | ID of the company (if applicable) |
| `days_in_waiting_list`       | Days spent on the waiting list |
| `customer_type`              | Customer classification |
| `adr`                        | Average Daily Rate (Euro) |
| `required_car_parking_spaces`| Number of parking spaces requested |
| `total_of_special_requests`  | Number of special requests |
| `reservation_status`         | Status of the booking (Canceled, Check-Out, No-Show) |
| `reservation_status_date`    | Date of status assignment |

```python
msno.matrix(df)
plt.title("Missing Data Matrix")
plt.show()
```

```python
df['children'].fillna(0, inplace=True)
df['agent'].fillna('NoAgent', inplace=True)
df['company'].fillna('NoCompany', inplace=True)
df['agent'] = df['agent'].astype(str)
df['company'] = df['company'].astype(str)
df['agent'] = LabelEncoder().fit_transform(df['agent'])
df['company'] = LabelEncoder().fit_transform(df['company'])

df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['weekend_ratio'] = df['stays_in_weekend_nights'] / (df['total_stay'] + 1)
df['is_family'] = np.where((df['children'] + df['babies']) > 0, 1, 0)
df['booking_change_rate'] = df['booking_changes'] / (df['lead_time'] + 1)
df['adr_per_person'] = df['adr'] / (df['total_guests'] + 1)
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
label_cols = ['hotel', 'arrival_date_month', 'meal', 'country', 'market_segment',
              'distribution_channel', 'reserved_room_type', 'assigned_room_type',
              'deposit_type', 'customer_type', 'reservation_status']
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

# CLASSIFICATION

```python
X = df.drop(['is_canceled', 'reservation_status_date'], axis=1)
y = df['is_canceled']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=537)
```

```python
models = {
    "LogReg": LogisticRegression(max_iter=2000),
    "RandomForest": RandomForestClassifier(n_estimators=100),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "NaiveBayes": GaussianNB(),
    "ExtraTrees": ExtraTreesClassifier(),
    "GradientBoost": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier()
}
```

```python
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    results.append((name, acc, auc))
    print(f"\n {name}\nAccuracy: {acc:.4f}\nROC AUC: {auc:.4f}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

## ROC Curve Comparison

```python
plt.figure(figsize=(12, 8))
for name, model in models.items():
    probs = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, probs)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score(y_test, probs):.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve Comparison of All Models")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```

# REGRESSION ANALYSIS

```python
X_reg = df[df['adr'] < 500].drop(['adr', 'is_canceled', 'reservation_status_date'], axis=1)
y_reg = df[df['adr'] < 500]['adr']
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(scaler.fit_transform(X_reg), y_reg, test_size=0.2, random_state=537)
```

```python
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)
y_pred_reg = lr.predict(X_test_reg)
```

```python
print("\nðŸ“ˆ REGRESSION RESULTS")
print("RÂ² Score:", r2_score(y_test_reg, y_pred_reg))
print("MAE:", mean_absolute_error(y_test_reg, y_pred_reg))
print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.5, edgecolor='k')
plt.xlabel("Actual ADR")
plt.ylabel("Predicted ADR")
plt.title("ADR: Actual vs Predicted")
plt.grid(True)
plt.show()
```
